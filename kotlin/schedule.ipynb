```kotlin
import kotlinx.datetime.format.Padding
import kotlinx.datetime.format.char
import kotlin.time.Duration

val format = LocalDateTime.Format {
  year()
  char('-')
  monthNumber(padding = Padding.ZERO)
  char('-')
  dayOfMonth()
  char(' ')
  hour(padding = Padding.ZERO)
  char(':')
  minute(padding = Padding.ZERO)
}
dfJson.add("duration") { format.parse(event_end).toInstant(TimeZone.UTC) - format.parse(event_start).toInstant(TimeZone.UTC) }
  .filter { get("duration") as Duration >= 30.minutes }
  .distinctBy { event_start }
  .select { name and event_start  and event_end and "duration" }
  .sortBy { event_start }
```
